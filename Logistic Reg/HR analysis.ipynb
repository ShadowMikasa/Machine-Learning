```python
df = pd.read_csv(r"C:\Users\rshib\Desktop\Krato\Linear Reg\HR_comma_sep.csv")
df
```

```python
X = df.drop(columns = "left")
X
```

```python
X = pd.concat([X, pd.get_dummies(X.loc[:,["Department","salary"]])], axis=1)
X
```

```python
model.fit(X_train, y_train)
```
